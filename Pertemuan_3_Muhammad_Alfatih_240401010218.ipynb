```python
import pandas as pd
import numpy as np
import requests

file_id = '1LfQWProB0VjWN5q8bKuRIgn-stULfIRo'
download_url = f'https://docs.google.com/uc?export=download&id={file_id}'

raw_housing_data = pd.read_csv(download_url)
print("Data mentah berhasil dimuat dari Google Drive!")
print(f"Dimensi Awal Dataset: {raw_housing_data.shape[0]} baris & {raw_housing_data.shape[1]} kolom")
```

```python
raw_housing_data.info()
display(raw_housing_data.describe())
print("\nJumlah Data Kosong Tiap Fitur:")
print(raw_housing_data.isnull().sum())
```

```python
raw_housing_data.drop_duplicates(inplace=True)
print(f"Dimensi Dataset Setelah Eliminasi Duplikat: {raw_housing_data.shape[0]} baris")
```

```python
raw_housing_data['kota'] = raw_housing_data['kota'].str.strip().str.title()
raw_housing_data['kondisi'] = raw_housing_data['kondisi'].str.strip().str.lower()
print("Proses penyelarasan string kota dan kondisi sukses.")
```

```python
nilai_tengah_luas = raw_housing_data['luas_m2'].median()
nilai_tengah_harga = raw_housing_data['harga_juta'].median()
nilai_sering_kamar = raw_housing_data['kamar'].mode()[0]

raw_housing_data['luas_m2'] = raw_housing_data['luas_m2'].fillna(nilai_tengah_luas)
raw_housing_data['harga_juta'] = raw_housing_data['harga_juta'].fillna(nilai_tengah_harga)
raw_housing_data['kamar'] = raw_housing_data['kamar'].fillna(nilai_sering_kamar)
print("Imputasi nilai kosong selesai dilakukan.")
```

```python
target_features = ['harga_juta', 'luas_m2', 'tahun_bangun']
for feature in target_features:
    kuartil_1 = raw_housing_data[feature].quantile(0.25)
    kuartil_3 = raw_housing_data[feature].quantile(0.75)
    rentang_iqr = kuartil_3 - kuartil_1
    batas_bawah = kuartil_1 - (1.5 * rentang_iqr)
    batas_atas = kuartil_3 + (1.5 * rentang_iqr)
    raw_housing_data[feature] = raw_housing_data[feature].clip(batas_bawah, batas_atas)

print("Penanganan nilai ekstrem (outlier) menggunakan metode IQR Fence telah diterapkan.")
```

```python
assert raw_housing_data.isnull().sum().sum() == 0, 'Deteksi: Masih terdapat data kosong!'
assert raw_housing_data.duplicated().sum() == 0, 'Deteksi: Masih terdapat data duplikat!'

print(f"Verifikasi Berhasil! Dimensi Akhir Dataset Bersih: {raw_housing_data.shape[0]} baris")
raw_housing_data.to_csv('housing_clean.csv', index=False)
print("File 'housing_clean.csv' berhasil diekspor.")
```
