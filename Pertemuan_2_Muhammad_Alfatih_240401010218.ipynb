```python
# =========================================================
# PERTEMUAN 2: Manajemen & Eksplorasi Data Tabular
# =========================================================
import pandas as pd
import numpy as np

# Mengunduh dataset utama Titanic langsung dari repositori publik
source_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data_titanic = pd.read_csv(source_url)

print("Proses import data berhasil!")
print(f"Dimensi Dataset: {data_titanic.shape[0]} entri baris dan {data_titanic.shape[1]} kolom fiturnya.")
```

```python
total_baris, total_kolom = data_titanic.shape
print(f"Ringkasan Data: Terdiri dari {total_baris} baris & {total_kolom} kolom.\n")

print("Daftar Fitur/Kolom Tersedia:")
for nomor, nama_kolom in enumerate(data_titanic.columns, 1):
    print(f"  [{nomor}] -> {nama_kolom}")

print("\nMenampilkan 5 Baris Data Teratas:")
display(data_titanic.head(5))

print("\nStruktur Tipe Data Komponen:")
data_titanic.info()
```

```python
jumlah_null = data_titanic.isnull().sum()
rasio_null   = (data_titanic.isnull().sum() / len(data_titanic)) * 100

tabel_kondisi_data = pd.DataFrame({
    'Total Kosong (Sells)': jumlah_null,
    'Persentase Hilang (%)': rasio_null.round(2)
})

kolom_bermasalah = tabel_kondisi_data[tabel_kondisi_data['Total Kosong (Sells)'] > 0].sort_values('Total Kosong (Sells)', ascending=False)
print(f"Hasil Analisis: Dari {data_titanic.shape[1]} kolom, ditemukan {len(kolom_bermasalah)} kolom yang tidak lengkap:\n")
print(kolom_bermasalah)
```

```python
jumlah_null = data_titanic.isnull().sum()
rasio_null   = (data_titanic.isnull().sum() / len(data_titanic)) * 100

tabel_kondisi_data = pd.DataFrame({
    'Total Kosong (Sells)': jumlah_null,
    'Persentase Hilang (%)': rasio_null.round(2)
})

kolom_bermasalah = tabel_kondisi_data[tabel_kondisi_data['Total Kosong (Sells)'] > 0].sort_values('Total Kosong (Sells)', ascending=False)
print(f"Hasil Analisis: Dari {data_titanic.shape[1]} kolom, ditemukan {len(kolom_bermasalah)} kolom yang tidak lengkap:\n")
print(kolom_bermasalah)
```

```python
status_selamat = data_titanic['Survived'].value_counts()
rate_selamat   = data_titanic['Survived'].mean() * 100

print(f"Jumlah Korban Selamat : {status_selamat[1]} jiwa")
print(f"Jumlah Korban Gugur   : {status_selamat[0]} jiwa")
print(f"Total Sampel Data     : {len(data_titanic)} orang")
print(f"\nRasio Akhir: Sebanyak {rate_selamat:.2f}% penumpang dinyatakan selamat, sedangkan {100 - rate_selamat:.2f}% lainnya tidak bertahan.")
```

```python
subset_wanita_vip = data_titanic[(data_titanic['Sex'] == 'female') & (data_titanic['Pclass'] == 1)]
total_wanita_vip = len(subset_wanita_vip)
rata_umur_vip = subset_wanita_vip['Age'].mean()
catatan_umur_valid = subset_wanita_vip['Age'].notna().sum()

print(f"Karakteristik Penumpang Wanita Kelas 1:")
print(f"-> Total Populasi: {total_wanita_vip} orang")
print(f"-> Rata-rata Usia: {rata_umur_vip:.2f} tahun (Dihitung dari {catatan_umur_valid} data umur yang terisi)")
```

```python
analisis_per_kelas = data_titanic.groupby('Pclass')['Survived'].mean() * 100
print("Persentase Kelangsungan Hidup Tiap Kelas Penumpang:\n")
for kelas_kamar, peluang in analisis_per_kelas.items():
    print(f"  * Kamar Kelas {kelas_kamar} : {peluang:.2f}% peluang selamat")

id_kelas_terbaik = analisis_per_kelas.idxmax()
print(f"\nKesimpulan: Penumpang di Kelas {id_kelas_terbaik} memegang rekor keselamatan tertinggi, yaitu sebesar {analisis_per_kelas[id_kelas_terbaik]:.2f}%.")
```

```python
data_titanic['KategoriUsia'] = pd.cut(
    data_titanic['Age'],
    bins=[0, 12, 18, 35, 55, 80],
    labels=['Fase Anak (0-12)', 'Fase Remaja (13-18)', 'Fase Produktif (19-35)', 'Fase Matang (36-55)', 'Fase Lansia (56-80)']
)

metrik_usia = data_titanic.groupby('KategoriUsia', observed=True).agg(
    Total_Manusia=('Survived', 'count'),
    Total_Selamat=('Survived', 'sum'),
    Rasio_Selamat_Pct=('Survived', 'mean')
).reset_index()

metrik_usia['Rasio_Selamat_Pct'] = (metrik_usia['Rasio_Selamat_Pct'] * 100).round(2)

print("Tabel Distribusi Keselamatan Berdasarkan Rentang Usia:\n")
print(metrik_usia.to_string(index=False))

print("\n\nInterpretasi Hasil:")
print("1. Golongan Anak (0-12 tahun) menempati urutan teratas dalam tingkat penyelamatan (~57.97%).")
print("2. Prioritas utama evakuasi tampaknya condong pada keselamatan anak-anak, sejalan dengan hukum laut universal.")
print("3. Sebaliknya, kaum Lansia (56-80 tahun) mencatatkan tingkat kelangsungan hidup paling minim, hanya menyentuh angka 30.00%.")

data_titanic.drop('KategoriUsia', axis=1, inplace=True)
```
